# Approximating a Sine Function

```python
# Necessary imports
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Create the sine wave that will be approximated
x = np.linspace(-np.pi, np.pi, 2000)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# Define the hyper parameters
lr       = 1e-6    #learning rate (determines size of step we take)
epochs   = 2000       #number of epochs (number of passes of entire training dataset)
```

```python
# Remember the equation of a polynomial: a + bx + cx**2 + dx**3 + ...
# Define the coeffecients of the polynomial that will be 'learned'
# In machine learning terms these are our 'weights'
a = np.random.rand()
b = np.random.rand()
c = np.random.rand() 
d = np.random.rand()
```

```python
# Iterate over the set number of epochs to 'learn' the weights
for i in range(epochs):
  # Calculate the y values for the weights in each iteration
  y_pred = a + b*x + c*x**2 + d*x**3 

  # Display the loss function every 100 epochs 
  # Simple loss function is equal to the sum of (y-y_pred)**2
  if i % 100 == 0:
    loss = np.mean(np.square(y-y_pred))
    print(loss)

  # Calculate the gradient of the loss function so we can use it to edit weights
  # (differentiate the loss function)
  grad = 2 * (y_pred - y)

  # Update each coefficient 
  # Update them by the learning rate * gradient
  a -= lr * np.sum(grad)
  b -= lr * np.sum(grad * x)
  c -= lr * np.sum(grad * x**2)
  d -= lr * np.sum(grad * x**3)

# Print out the final equation
print(f'Result: y = {a} + {b} x + {c} x^2 + {d} x^3')
```

```python
# Calculate the predicted output using these trained coefficients
y_pred = a + b*x + c*x**2 + d*x**3
```

```python
# Plot the original sine wave and the predicted sine wave
# Going back and changing the epochs and lr will affect the result 
plt.plot(x,y)
plt.plot(x,y_pred)
plt.show()
```
